```python
###Regras de pontuação de postos###

#Verificar outros campos com 'NaN' e outros alterações
df_postos['Possui combustível aditivado?'] = df_postos['Possui combustível aditivado?'].fillna("Não")
df_postos['Ofereceu abastecer com gasolina aditivada ou etanol aditivado?'] = df_postos['Ofereceu abastecer com gasolina aditivada ou etanol aditivado?'].fillna("Não")
df_postos['Movimento na pista (Carros e Frentistas):'] = df_postos['Movimento na pista (Carros e Frentistas):'].astype(int)
df_postos.drop(columns=['Anexo do comprovante'], inplace=True)
df_postos['Carimbo de data/hora'] = df_postos['Carimbo de data/hora'].astype('datetime64[D]')

df_postos['1'] = np.where(df_postos['Aguardou na postura adequada e sinalizou. Ou estava atendendo mais de um carro?'] == "Sim", 10, 0)
df_postos['2'] = np.where(df_postos['O funcinoário estava devidamente uniformizado, inclusive com crachá?'] == "Sim", 10, 0)
df_postos['3'] = np.where(df_postos['Cumprimentou com simpatia, dizendo "Bom dia", "Boa tarde" ou "Boa noite"?'] == "Sim", 10, 0)
df_postos['4'] = np.where(df_postos['Possui combustível aditivado?'] == "Não",0, np.where(df_postos['Ofereceu abastecer com gasolina aditivada ou etanol aditivado?'] == "Sim", 20, 0))
df_postos['5'] = np.where(df_postos['Possui combustível aditivado?'] == "Não", 
                                                                                                                                                                                                                   np.where(df_postos['Cumpriu os procedimentos básicos, (deixou  o bico de abastecimento no automático e ofereceu os serviços de cortesia "radiador", "vidros" e "pneus")? Ou estava atendendo mais de um carro?'] == "Sim", 30, 0), 
                                                                                                                                                                                                                   np.where(df_postos['Cumpriu os procedimentos básicos, (deixou  o bico de abastecimento no automático e ofereceu os serviços de cortesia "radiador", "vidros" e "pneus")? Ou estava atendendo mais de um carro?'] == "Sim", 20, 0))
df_postos['6'] = np.where(df_postos['Possui combustível aditivado?'] == "Não", np.where(df_postos['Sinalizou para abrir o capô e ofereceu um produto adicional? Ou estava atendendo mais de um carro?'] == "Sim" , 20,0), np.where(df_postos['Sinalizou para abrir o capô e ofereceu um produto adicional? Ou estava atendendo mais de um carro?'] == "Sim" , 10, 0))
df_postos['7'] = np.where(df_postos['Possui combustível aditivado?'] == "Não", np.where(df_postos['Os banheiros estavam Limpos?'] == "Sim" , 20, 0), np.where(df_postos['Os banheiros estavam Limpos?'] == "Sim" , 10, 0))
df_postos['8'] = np.where(df_postos['Agradeceu, se despediu?'] == "Sim", 10, 0)
df_postos['9'] = np.where(df_postos['Convidou a voltar?'] == "Sim", 10, 0)

#Totais de pontos
df_postos['Total de Pontos'] = (
  df_postos['1'] + 
  df_postos['2'] +
  df_postos['3'] + 
  df_postos['4'] + 
  df_postos['5'] + 
  df_postos['6'] + 
  df_postos['7']  + 
  df_postos['8'] + 
  df_postos['9'])

#Deletar colunas utilizadas para soma
df_postos.drop(columns=['1', '2', '3', '4', '5','6', '7', '8','9'], inplace=True)
```

```python
###Regras de pontuação de conveniencia###

#Verificar outros campos com 'NaN' e outras alterações
df_conveniencia['Quantidade de clientes na conveniência:'] = df_conveniencia['Quantidade de clientes na conveniência:'].astype(int)
df_conveniencia.drop(columns=['Anexo do comprovante'], inplace=True)
df_conveniencia['Carimbo de data/hora'] = df_conveniencia['Carimbo de data/hora'].astype('datetime64[D]')

df_conveniencia['1'] = np.where(df_conveniencia['Cumprimentou com simpatia dizendo "Bem Vindo a Rede Duque"?'] == "Sim", 20, 0) 
df_conveniencia['2'] =  np.where(df_conveniencia['O funcionário estava devidamente uniformizado?'] == "Sim", 10, 0)
df_conveniencia['3'] =  np.where(df_conveniencia['Food- Ofereceu a promoção?'] == "Sim", 20, 0)
df_conveniencia['4'] =  np.where(df_conveniencia['Entregou a comanda preenchida antes do cliente dirigir-se ao caixa? Caso o cliente tenha pedido café expresso, foi servido água com gás e cookie? Foi entregue 01 sache de acordo com a escolha do cliente?']  == "Sim", 10, 0)
df_conveniencia['5'] =  np.where(df_conveniencia['Ofereceu NF paulista, registrou a venda, imprimiu o cupom?'] == "Sim", 10, 0)
df_conveniencia['6'] =  np.where(df_conveniencia['Agradeceu, se despediu e convidou a voltar?'] == "Sim", 10, 0)
df_conveniencia['7'] =  np.where(df_conveniencia['Os banheiros estavam Limpos?'] == "Sim", 20, 0)
df_conveniencia['8'] =  np.where(df_conveniencia['Estava usando o celular?'] == "Sim", 0, 10)

#Totais de pontos
df_conveniencia['Total de Pontos']  = (
  df_conveniencia['1'] +
  df_conveniencia['2'] +
  df_conveniencia['3'] +
  df_conveniencia['4'] +
  df_conveniencia['5'] +
  df_conveniencia['6'] +
  df_conveniencia['7'] +
  df_conveniencia['8'])

#Deletar colunas utilizadas para soma
df_conveniencia.drop(columns=['1', '2', '3', '4', '5','6', '7', '8'], inplace=True)
```

```python
###Regras de pontuação de lava-rapido###

#Verificar outros campos com 'NaN' e outras alterações
df_lava_rapido['Quantos carros tinham na fila da lavagem?'] = df_lava_rapido['Quantos carros tinham na fila da lavagem?'].astype(int)
df_lava_rapido.drop(columns=['Anexo do comprovante'], inplace=True)
df_lava_rapido['Carimbo de data/hora'] = df_lava_rapido['Carimbo de data/hora'].astype('datetime64[D]')

df_lava_rapido['1'] = np.where(df_lava_rapido['Postura adequada, estava devidamente uniformizado, inclusive com crachá?'] == "Sim", 10 ,0)
df_lava_rapido['2'] = np.where(df_lava_rapido['Cumprimentou com simpatia, dizendo "Bom dia", "Boa tarde" ou ""Boa noite?'] == "Sim", 10, 0)
df_lava_rapido['3'] = np.where(df_lava_rapido['Ofereceu Lavar com Cera?'] == "Sim", 30, 0)
df_lava_rapido['4'] = np.where(df_lava_rapido['Cumpriu os procedimentos básicos ? (Lavou o carro na parte da lataria com luva  e as rodas e pneus com outra luva)'] == "Sim", 20, 0)
df_lava_rapido['5'] = np.where(df_lava_rapido['No término da lavagem, perguntou para o cliente se o carro ficou bem lavado?'] == "Sim", 30, 0)
df_lava_rapido['6'] = np.where(df_lava_rapido['Parte interna do carro ficou limpa?'] == "Sim", 10, 0)
df_lava_rapido['7'] = np.where(df_lava_rapido['Parte externa do carro ficou limpa?'] == "Sim", 10, 0)
df_lava_rapido['8'] = np.where(df_lava_rapido['Agradeceu, se despediu e convidou a voltar?'] == "Sim", 10, 0)
df_lava_rapido['9'] = np.where(df_lava_rapido['Quanto tempo demorou para lavar o carro? (Minutos)'] <= 60, 10, 0)

#Totais de pontos
df_lava_rapido['Total de Pontos'] = (
  df_lava_rapido['1'] +
  df_lava_rapido['2'] +
  df_lava_rapido['3'] +
  df_lava_rapido['4'] +
  df_lava_rapido['5'] +
  df_lava_rapido['6'] +
  df_lava_rapido['7'] +
  df_lava_rapido['8'] +
  df_lava_rapido['9'])

#Deletar colunas utilizadas para soma
df_lava_rapido.drop(columns=['1', '2', '3', '4', '5','6', '7', '8', '9'], inplace=True)
```

```python
#Fazer JOIN com supervisores e diretores

#Dividir datasets por tipo de serviço
df_supervisores_postos = df_supervisores.loc[df_supervisores['TIPO SERVIÇO'] == 'POSTO']
df_supervisores_lava_rapido = df_supervisores.loc[df_supervisores['TIPO SERVIÇO'] == 'LAVA-RÁPIDO']
df_supervisores_conveniencia = df_supervisores.loc[df_supervisores['TIPO SERVIÇO'] == 'CONVENIÊNCIA']

df_supervisores_postos = df_supervisores_postos.rename(columns = {'ESTABELECIMENTO': 'Posto'})
df_supervisores_lava_rapido = df_supervisores_lava_rapido.rename(columns = {'ESTABELECIMENTO': 'Posto'})
df_supervisores_conveniencia = df_supervisores_conveniencia.rename(columns = {'ESTABELECIMENTO': 'Posto'})

#Fazer join e criar coluna em cada dataset
df_supervisores_postos = pd.merge(df_postos, df_supervisores_postos, how="left", on='Posto')
df_supervisores_lava_rapido = pd.merge(df_lava_rapido, df_supervisores_lava_rapido, how="left", on='Posto')
df_supervisores_conveniencia = pd.merge(df_conveniencia, df_supervisores_conveniencia, how="left", on='Posto')
```
